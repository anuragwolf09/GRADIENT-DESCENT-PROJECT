THEREFORE HERE OUR TARGET IS TO FIND M AND C USING GRADIENT DESCENT APPROACH AND THEN USING THAT M AND C WE WILL PREDICT ON TEST DATASET.

```python
x_train=dtrain[:,0:13]
y_train=dtrain[:,13]
```

```python
dtest=np.loadtxt(r'C:\Users\hp\Documents\0000000000002417_test_boston_x_test.csv',delimiter=',')
dtest.shape
```

```python
from sklearn import preprocessing as pp
scaled=pp.StandardScaler()
scaled.fit(x_train)
x_train=scaled.transform(x_train)
```

```python
dtest=scaled.transform(dtest)
```

```python
y_train
```

```python
def step_gd(x, y, learning_rate, m):
    M = x.shape[0]
    m_slope = (-2/M)*(((y - (x*m).sum(axis=1)).reshape(-1, 1))*x).sum(axis=0)
    m = m - learning_rate * m_slope
    return m
```

```python
def gd(x, y, learning_rate, num_iterations):
    m = np.zeros(x.shape[1])
    for i in range(num_iterations):
        m = step_gd(x, y, learning_rate, m)
        print(i, "Cost :", cost(x, y, m))
    return m
```

```python
def cost(x,y,m):
    return ((y - (x*m).sum(axis=1)) ** 2).mean()
```

```python
def run(x,y):
    df=pd.DataFrame(x)
    df[x.shape[1]]=1
    x=df.values
    learning_rate=0.009
    num_iterations=500
    m=gd(x,y,learning_rate,num_iterations)
    return m[0:13],m[13]
    
```

```python
m,c=run(x_train,y_train)
print(m,c)
```

```python
def predict(x,m,c):
    n=len(x)
    y_pred=np.zeros(n)
    for i in range(n):
        X=x[i,:]
        y_pred[i]=(m*X).sum()+c
        
    return y_pred
```

```python
y_predict=predict(dtest,m,c)
```

```python
y_predict.mean()
```

```python
len_y=len(y_predict)
for i in range(len_y):
    print(y_predict[i])
```

```python
def step_gd(dtrain,learning_rate,m):
    M=len(dtrain)
    mslope=np.zeros(14)
    for i in range(M):
        x=dtrain[i,0:13]
        x=np.append(x,1)
        y=dtrain[i,13]
        for j in range(14):
            mslope[j]=mslope[j]+(-2/M)*(y - (m * x).sum())*x[j]
        m=m-(learning_rate*mslope)
    return m
        
```

```python
def gd(dtrain,learning_rate,num_iterations):
    m=np.zeros(14)
    for i in range(num_iterations):
        m=step_gd(dtrain,learning_rate,m)
        print(i,'cost:',cost(dtrain,m))
    return m
```

```python
def cost(dtrain,m):
    total_cost=0
    M=len(dtrain)
    for i in range(M):
        x=dtrain[i,0:13]
        x=np.append(x,1)
        y=dtrain[i,13]
        total_cost+=(1/M)*((y - (m*x).sum() )**2)
    return total_cost
        
```

```python
def run():
    learning_rate=0.001
    num_iterations=100
    m=gd(dtrain,learning_rate,num_iterations)
    return m[0:13],m[13]
    
```

```python
m,c=run()
print(m,c)
```
