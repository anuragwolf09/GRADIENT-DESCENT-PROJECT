```python
def step_gd(x_train,y_train,learning_rate,m):
    M=x_train.shape[0]
    x=x_train
    y=y_train
    m_slope = (-2/M)*(((y - (x*m).sum(axis=1)).reshape(-1, 1))*x).sum(axis=0)
    m = m - learning_rate * m_slope
    return m
    
```

```python
def predict(x,m,c):
    n=len(x)
    y_pred=np.zeros(n)
    for i in range(n):
        X=x[i,:]
        y_pred[i]=(m*X).sum()+c
        
    return y_pred
```
